### ***FIGURA (a)***

```python
from numpy.core.numeric import indices
#Este dataframe del cuadro (a) tiene 3 columnas,ya que es una columna por cada línea (Función)
Numero_moleculas=[[3,7,5],[0,2,4],[0,0,1],]
velocidad=["500","1000","1500"]
Temperatura=["100","300","700"]
df=pd.DataFrame(Numero_moleculas, index=velocidad, columns=Temperatura)
df
#Como no tenemos datos acerca del número de moléculas, entonces los tomamos como números enteros ascendentes al eje "y".
```

Estas gráficas corresponden a las distribuciones de Maxwell Boltzmann por lo que podemos usar su teoría para explicar el problema.

El pico más alto en cada Función corresponde a la moda, es decir, que la mayoría de partículas de 100k tendrá menor velocidad que el pico de 300k y cuando el N2 tiene 700k la mayoría de partículas tiene mayor velocidad que en 300k.

```python
import seaborn as sns
import matplotlib.pyplot as plt
x=["100k","300k","700k"]
y=[250,350,550]          # Aproximamos las velocidades en los picos de forma práctica para su explicación
plt.bar(x, y)
plt.show()
```

Aquí se puede observar mejor como la mayoría de partículas en 700k tienen mayor velocidad, dado que tambien tienen mayor Energía Cinética.

### ***FIGURA (b)***


Dado que la tempratura para todos los gases es la misma, la velocidad está relacionada de manera directa con su peso melecular.Donde las moléculas más ligeras tienden a tener una moda de velocidad mayor frente a las moléculas más pesadas.


```python
import matplotlib.pyplot as plt
```

```python
moleculas=["Cl2","N2","He"]
velocidad=[[250],[450],[1025]]
df=pd.DataFrame(velocidad,index=moleculas)
df
```

```python

gases="Cl2","N2","He"
velocidad=[250,450,1025]
separa=(0.1,0.1,0.1)
plt.pie(velocidad, labels=gases, explode=separa,shadow=True,autopct="%1.1f%%")
plt.show
```

En esta gráfica las regiones corresponen al número de moléculas de cada gas, se puede diferenciar facilmente que la mayoría de moléculas en el He les corresponde una mayor velocidad en general.
